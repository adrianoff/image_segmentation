```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('image.tif')
kernel = np.array([[-0.1,-0.1,-0.1], [-0.1,3,-0.1], [-0.1,-0.1,-0.1]])
im = cv2.filter2D(img, -1, kernel)
plt.imshow(im)
plt.show()
cv2.imwrite('test3.tif', im)
```

```python
gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
plt.imshow(gray)
plt.show()
cv2.imwrite('test3_gray.tif', gray)
```

```python
gray = cv2.GaussianBlur(gray,(5,5),0)
plt.imshow(gray)
plt.show()
cv2.imwrite('test3_gray.tif', gray)
```

```python
th3 = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 13, 2)
plt.imshow(th3)
plt.show()
cv2.imwrite('test3_countours.tif', th3)
```
